```python
#we can also draw a picture or graph inside another graph
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8]) #big values in axes
axes2=fig.add_axes([0.2,0.5,0.4,0.3]) #small values in axes
axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

```python
from matplotlib.pylab import plt
fig,axes=plt.subplots()
axes.set_title('title')
axes.set_xlabel('x')
axes.set_ylabel('y')
#axes.plot(x,y,'r')
```

```python
fig,axes=plt.subplots(dpi=200) #if we require we put dpi otherwise no
axes.set_xlabel("x_label")
axes.set_ylabel("y_label")
axes.set_title("title")
axes.plot(x,x**2)
axes.plot(x,x**3)
```

```python
axes.legend(["y=x**2","y=x**3"],loc=3)
```

```python
#in matplotlib, you can set other property such as the color transparency and more
fig,axes=plt.subplots(dpi=100)
axes.plot(x,x+1,color='red',alpha=.5,lw=1.50,linestyle='-',marker='o',markersize=5,markerfacecolor='blue') #only values is (x,x+1) are conform ,and reamaing are all optional
axes.plot(x-1,x+2,color='blue',lw=5.50,linestyle='-',marker='*',markersize=5,markerfacecolor='blue') #like this
axes.plot(x,x+4) #like this remaing are all optional
```
