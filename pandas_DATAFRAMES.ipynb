#create dataframe with dictionary arry

```python
data={'animal':['cat','cat','snake','dog','dog'],
      'age':[2.5,3,np.nan,5,2],
      'visits':[1,3,2,3,2],
      'priority':['yes','yes','no','yes','no']}
labels=['a','b','c','d','e']
df2=pd.DataFrame(data,index=labels)
df2
```

```python
print(df2.head(4))
print(df2.head(-1))
print(df2.head(1))
print(df2.head(4))
print(df2.tail(3))
print(df2.tail())
print(df2.index)
print(df2.columns)
print(df2.values)
print(df2.T)
print(df2.describe())
```
