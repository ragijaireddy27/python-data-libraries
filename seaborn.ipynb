2nd plot 

PairGrid

```python
g=sns.PairGrid(iris)
g=g.map_upper(sns.scatterplot)
g=g.map_diag(sns.kdeplot,lw=2)
g=g.map_lower(sns.kdeplot,color='red')
```

```python
g=sns.PairGrid(iris,hue='species',palette='plasma')
g=g.map_upper(sns.scatterplot,lw=2,edgecolor='w',s=45)
g=g.map_diag(sns.histplot)
g=g.map_lower(sns.kdeplot)
g=g.add_legend()
```
