```python
import matplotlib.pylab as plt
import numpy as np
```

```python
x=np.linspace(start=1,stop=100,num=20)
#x=np.linspace(1,100,20) # here 20 tottal number b/w 1 to 100
y=2*x+3
```

```python
plt.plot(x,y,label='y=2*x+3') # in the place of polt we can use bar,line,piechart diifeerent graphs alaso
plt.legend() # if above label show in fig we need call legend() compalsary 
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ploting the linear equation')
plt.xlim(0,10) #lower limit 0,upper limit 10 in x axis
plt.ylim(0,20)#lower limit 0,upper limit 20 in y axis
plt.grid(True) #that grid background brackts
plt.savefig('linear -figure.png') # for saving the figure in our local system
plt.show()
```

```python
fig=plt.figure() # it is empty paper og figure we can draw any thing in that figure
#plt.figure(figsize=(220,299) if we need we provide the size of the figuree
ax=fig.add_axes([0.5,0.8,0.7,0.6]) # there are four arguments it is compalsary (lft,bottom,width,lenght)
# we can plot values in that block or figure or that empty paper
x=np.linspace(1,20,20)
y=np.sin(x)
ax.plot(x,y) #we provide label also and colour and style,also ex(ax.plot(x,y,'r') or plot(x,y,label='y=sin x','r'
```

```python
#generate x values
x=np.linspace(1,100,15)
#create blank canvas or empty sheet
fig=plt.figure(figsize=(19,8))

#to plot on center main plot
axes=fig.add_axes([0.3 ,0.5,0.7,0.9])
axes.plot(x,x+1)
axes.set_title('center plot')
#top plot -summary
axes=fig.add_axes([0.1,0.75,0.75,0.5])
axes.plot(x,2*x+1)

axes.set_title('top plot')
#right plot -auxiliary
axes=fig.add_axes([0.75,0.3,0.3,0.4])
axes.plot(x,0.5*x+1)

axes.set_title('right plot')
#left plot contextual information
axes=fig.add_axes([0.05,0.3,0.2,0.4])
axes.plot(x,3*x+2)

axes.set_title('left plot')
#bottom plot -comparative data
axes=fig.add_axes([0.3,0.05,0.4,0.2])
axes.plot(x,-x+1)
axes.set_title('bottom')
#top right corner
axes=fig.add_axes([0.5,0.5,0.7,0.9])
axes.plot(x,4*x+1)
axes.set_title('top right corner')
```

```python
# creating sub plot it is very use ful than figure because in figure we plot only on axies but in subplot we can plot more than one axes
fig,ax=plt.subplots(1,1)
ax.plot(x,x+1)
ax.plot(x,x**2)
ax.plot(x,x*100)
```

```python
# we can pass parameter like
fig,ax=plt.subplots(2,2,figsize=(10,3),dpi=100,sharey=True,subplot_kw={'facecolor':'lightgray'})

# we can use different grphs in single graphs
ax[0,0].plot(x,x+1)
ax[0,1].plot(x,x*2)
ax[1,0].plot(x,x+100)

ax[1,1].plot(x,x**2)
```

```python
# there are diffrent style if we want any style then we want that name that is avilable in matplo.com website
#now we saw two example
#1st style
plt.style.use('tableau-colorblind10')
plt.plot([1,2,3,4],[1,4,8,17])
plt.show()
#2nd example of style
plt.style.use('ggplot')
plt.plot([10,20,30,40],[1,4,8,17])
plt.show()
```
